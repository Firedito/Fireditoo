Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
heart_data = pd.read_csv('/content/heart.csv')
```

Splitting the Features and Target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

Langkah EDA dengan Visualisasi Data

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi distribusi target
sns.countplot(x=Y)
plt.title("Distribusi Target")
plt.show()
```

```python
# Heatmap korelasi antar fitur
correlation_matrix = X.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Heatmap Korelasi Antar Fitur")
plt.show()
```

```python
# Visualisasi distribusi beberapa fitur penting
for column in X.columns[:4]:  # Contoh hanya 4 fitur pertama
    sns.histplot(X[column], kde=True)
    plt.title(f"Distribusi {column}")
    plt.show()
```

 Normalisasi/Standarisasi Dataset

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split ulang data setelah normalisasi
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, stratify=Y, random_state=2)
```

Benchmarking dengan Algoritma Lain

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier

# Random Forest
rf_model = RandomForestClassifier(random_state=2)
rf_model.fit(X_train, Y_train)
rf_accuracy = rf_model.score(X_test, Y_test)
print("Random Forest Accuracy:", rf_accuracy)
```

```python
# KNN
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, Y_train)
knn_accuracy = knn_model.score(X_test, Y_test)
print("KNN Accuracy:", knn_accuracy)
```

```python
# Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=2)
gb_model.fit(X_train, Y_train)
gb_accuracy = gb_model.score(X_test, Y_test)
print("Gradient Boosting Accuracy:", gb_accuracy)
```

Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm # Import the 'svm' module

# Contoh untuk SVM
param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
grid_search = GridSearchCV(svm.SVC(probability=True), param_grid, scoring='accuracy', cv=5) # Now 'svm' is defined
grid_search.fit(X_train, Y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Score:", grid_search.best_score_)
```

```python
# Model dengan parameter terbaik
best_model = grid_search.best_estimator_
test_accuracy = best_model.score(X_test, Y_test)
print("Test Accuracy with Best Parameters:", test_accuracy)
```
